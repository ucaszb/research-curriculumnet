```python
%matplotlib inline
import sys
import numpy as np
import magpy as mp
import matplotlib.pyplot as plt
from math import sin, cos, asinh, atan, sqrt, pi
```

```python
mode = ['yigeskx', 'sangeskx', 'sanjiaoskx', 'liujiaoskx', 'haoduoskx', 'biaozhun']
```

```python
M=mp.MagTEM(n_system=(256, 256, 25), dx=(4e-09, 4e-09, 4e-09))
M.zb_skx(period=200e-09, origin_offset=[128,128], mode=mode[5], num=7)
M.show_m(skip=5)
```

```python
phase=M.calculate_phase()
c=mp.show_2D_scalar_field(np.abs(phase), ticks_scale=M.dx)
plt.colorbar(c, label='Phase (rad)')
plt.show()
```

```python
defocus_list = [100e-6, 0, -100e-6]
defocus_label_list = ['under', 'in-focus', 'over']
imgs = M.defocus_series(phase=phase, dz_list=defocus_list)
```

```python
idx = 0
plt.figure(figsize=(48,48))
ax = plt.subplot(1, 1, 1)
ax.imshow(imgs[idx].T, cmap=plt.cm.gray, vmin=0.75, vmax=1.25)
ax.set_title(defocus_label_list[idx])
ax.invert_yaxis()
ax.axis('off')
plt.show()
```
