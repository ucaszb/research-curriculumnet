```python
%matplotlib inline
import sys
import numpy as np
import magpy as mp
import matplotlib.pyplot as plt
```

### SkX_scaled

```python
M=mp.MagTEM(n_system=(256, 256, 25), dx=(4e-09, 4e-09, 4e-09))
M.SkX_scaled(period=128e-9)
M.show_m(skip=5)
```

```python
phase=M.calculate_phase()
c=mp.show_2D_scalar_field(np.abs(phase), ticks_scale=M.dx)
plt.colorbar(c, label='Phase (rad)')
plt.show()
```

```python
defocus_list = [50e-6, 0, -50e-6]
defocus_label_list = ['under', 'in-focus', 'over']
imgs = M.defocus_series(phase=phase, dz_list=defocus_list)

fig, axes = plt.subplots(1, len(defocus_list))
for ax, df, img, defocus_label in zip(axes.reshape(-1), defocus_list, imgs, defocus_label_list):
    ax.imshow(img.T, cmap=plt.cm.gray, vmin=0.75, vmax=1.25)
#    ax.set_title(str(df*1e6)+' $\mu$m')
    ax.set_title(defocus_label)
    ax.invert_yaxis()
    ax.axis('off')
plt.show()
```
